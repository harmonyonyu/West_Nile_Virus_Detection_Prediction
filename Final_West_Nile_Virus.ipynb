```python
# There will be an issue here with the codes with '+' appeneded codes(ex.'FG+'). 
# Remember to follow Will's recomendation to get FG and FG+ and then go back and remove from
# FG row all '1s' where '1s' are in 
# Codes: +FC, FC, TS, GR, RA, DZ, SN, SG, GS, PL, IC, FG+, FG, BR, UP, HZ, FU, VA, DU, DS,
# PO, SA, SS, PY, SQ, DR, SH, FZ, MI, PR, BC, BL, VC
# This long block of code creates a list for every code type.  If the code is present in the 
# row the function will append a 1 to the list, otherwise a 0

#NOTE TO SELF, DO A FURTHER INVESTIGATION TO SEE IF "+" or "-" is anywhere else in the codes

FC_plus = []
FG_plus = []
FC = []
TS = []
GR = []
RA = []
DZ = []
SN = []
SG = []
GS = []
PL = []
IC = []
FG = []
BR = []
UP = []
HZ = []
FU = []
VA = []
DU = []
DS = []
PO = []
SA = []
SS = []
PY = []
SQ = []
DR = []
SH = []
FZ = []
MI = []
PR = []
BC = []
BL = []
VC = []

def codesum (code):
    for i in code:
        if '+FC' in i:
            FC_plus.append(1)
        else:
            FC_plus.append(0)

    for i in code:
        if 'FC' in i:
            FC.append(1)
        else:
            FC.append(0)
            
    for i in code:
        if 'TS' in i:
            TS.append(1)
        else:
            TS.append(0)
            
    for i in code:
        if 'GR' in i:
            GR.append(1)
        else:
            GR.append(0)
                
    for i in code:
        if 'RA' in i:
            RA.append(1)
        else:
            RA.append(0) 

    for i in code:
        if 'DZ' in i:
            DZ.append(1)
        else:
            DZ.append(0) 
            
    for i in code:
        if 'SN' in i:
            SN.append(1)
        else:
            SN.append(0)
            
    for i in code:
        if 'SG' in i:
            SG.append(1)
        else:
            SG.append(0) 
            
    for i in code:
        if 'GS' in i:
            GS.append(1)
        else:
            GS.append(0)
            
    for i in code:
        if 'PL' in i:
            PL.append(1)
        else:
            PL.append(0) 
            
    for i in code:
        if 'IC' in i:
            IC.append(1)
        else:
            IC.append(0)
            
    for i in code:
        if 'FG+' in i:
            FG_plus.append(1)
        else:
            FG_plus.append(0) 
            
    for i in code:
        if 'FG' in i:
            FG.append(1)
        else:
            FG.append(0) 
            
    for i in code:
        if 'BR' in i:
            BR.append(1)
        else:
            BR.append(0)

    for i in code:
        if 'UP' in i:
            UP.append(1)
        else:
            UP.append(0)
   
    for i in code:
        if 'HZ' in i:
            HZ.append(1)
        else:
            HZ.append(0)
            
    for i in code:
        if 'FU' in i:
            FU.append(1)
        else:
            FU.append(0)
    
    for i in code:
        if 'VA' in i:
            VA.append(1)
        else:
            VA.append(0)    
            
    for i in code:
        if 'DU' in i:
            DU.append(1)
        else:
            DU.append(0)   
            
    for i in code:
        if 'DS' in i:
            DS.append(1)
        else:
            DS.append(0) 
            
    for i in code:
        if 'PO' in i:
            PO.append(1)
        else:
            PO.append(0)
            
    for i in code:
        if 'SA' in i:
            SA.append(1)
        else:
            SA.append(0) 
            
    for i in code:
        if 'SS' in i:
            SS.append(1)
        else:
            SS.append(0) 
            
    for i in code:
        if 'PY' in i:
            PY.append(1)
        else:
            PY.append(0) 
            
    for i in code:
        if 'SQ' in i:
            SQ.append(1)
        else:
            SQ.append(0) 
    
    for i in code:
        if 'DR' in i:
            DR.append(1)
        else:
            DR.append(0)   
            
    for i in code:
        if 'SH' in i:
            SH.append(1)
        else:
            SH.append(0) 
    
    for i in code:
        if 'FZ' in i:
            FZ.append(1)
        else:
            FZ.append(0)
            
    for i in code:
        if 'MI' in i:
            MI.append(1)
        else:
            MI.append(0)
            
    for i in code:
        if 'PR' in i:
            PR.append(1)
        else:
            PR.append(0)
            
    for i in code:
        if 'BC' in i:
            BC.append(1)
        else:
            BC.append(0)

    for i in code:
        if 'BL' in i:
            BL.append(1)
        else:
            BL.append(0)
            
    for i in code:
        if 'VC' in i:
            VC.append(1)
        else:
            VC.append(0)
```

```python
# CREATE TWO NEW COLUMNS 'WNV_SPECIES_TRUE', 'WNV_SPECIES_FALSE' 
# CREATE LIST OF WNV SPECIES TRUE BUGS FROM WNVPRESENT = 1 LIST 
# WNV_BUGS = ['CULEX PIPIENS/RESTUANS', 'CULEX PIPIENS', 'CULEX RESTUANS']
# ITERATE THROUGH SPECIES COLUMN 
# IF BUG IN WNV_BUGS 
### WNV_SPECIES_TRUE.APPEND 1 AND WNV_SPECIES_FALSE.APPEND 0
# ELSE 
### WNV_SPECIES_TRUE.APPEND 0 AND WNV_SPECIES_FALSE.APPEND 1 

wnv_species_true = []
wnv_species_false = []
wnv_bugs = ['CULEX PIPIENS/RESTUANS', 'CULEX PIPIENS', 'CULEX RESTUANS']

for bug in total_df.Species: 
    if bug in wnv_bugs: 
        wnv_species_true.append(1)
        wnv_species_false.append(0)
    else: 
        wnv_species_true.append(0)
        wnv_species_false.append(1)
        
df_wnv_species_true = pd.DataFrame(wnv_species_true, columns=['WNV_Species_True'], dtype='int')
df_wnv_species_false = pd.DataFrame(wnv_species_false, columns=['WNV_Species_False'], dtype='int')
```
